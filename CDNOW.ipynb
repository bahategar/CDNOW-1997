```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from xgboost import XGBClassifier, XGBRegressor
from sklearn.model_selection import GridSearchCV
```

```python
columns = ['customer_id', 'date', 'quantity', 'price']

df = pd.read_csv('data/CDNOW_master.txt', sep='\s+', names=columns)

print(df.info())
df.head()
```

```python
# Modify columns
df['date'] = df['date']\
                .astype('str')\
                .apply(lambda x: x[4:6] + '-' + x[6:len(x)] + '-' + x[:4])
df['date'] = pd.to_datetime(df['date'], format='%m-%d-%Y')
print(df.info())
df.head()
```

```python
# Visualize All purchase within cohort

df.reset_index()\
    .set_index('date')[['price']]\
    .resample(rule='MS')\
    .sum()\
    .plot()

plt.show()
```

```python
# Visualize Individual Customer Purchases

ids = df['customer_id'].unique()
ids_selected = ids[:10]

per_customer_df = df[df['customer_id'].isin(ids_selected)]\
                    .groupby(['customer_id', 'date'])\
                    .sum()\
                    .reset_index()

per_customer_df
```

```python
# Time splitting 
n_days = 90
max_date = df['date'].max()
cutoff = max_date - pd.to_timedelta(n_days, unit='d')

temp_in = df[df['date'] <= cutoff]
temp_out = df[df['date'] > cutoff]

```

```python
# Feature engineering (RFM)

targets = temp_out\
            .drop('quantity', axis=1)\
            .groupby('customer_id')\
            .agg({'price': sum})\
            .rename({'price': 'spend_90_total'}, axis=1)\
            .assign(spend_90_flag = 1)

targets
```

```python
# Recency

recency_feature = temp_in[['customer_id', 'date']]\
                    .groupby('customer_id')\
                    .apply(lambda x: (x['date'].max() - max_date) / pd.to_timedelta(1, 'day'))\
                    .to_frame()\
                    .set_axis(['recency'], axis=1)

recency_feature
```

```python
# Frequency

frequency_feature = temp_in[['customer_id', 'date']]\
                    .groupby('customer_id')\
                    .count()\
                    .set_axis(['frequency'], axis=1)

frequency_feature
```

```python
# monetary

monetary_feature = temp_in\
                    .groupby('customer_id')\
                    .agg({ 'price': ['sum', 'mean'] })\
                    .set_axis(['monetary_sum', 'monetary_mean'], axis=1)

monetary_feature
```

```python
# Combine feature

features = pd.concat([recency_feature, frequency_feature, monetary_feature], axis=1)\
            .merge(targets,
                  left_index=True,
                  right_index=True,
                  how='left')\
            .fillna(0)

features
```

```python
# Machine learning (regression)

from xgboost import XGBClassifier, XGBRegressor

from sklearn.model_selection import GridSearchCV

X = features[['recency', 'frequency', 'monetary_sum', 'monetary_mean']]
y_spending = features['spend_90_total']

xgb_reg = XGBRegressor(objective='reg:squarederror', random_state=123)

xgb_reg_model = GridSearchCV(estimator=xgb_reg, 
                             param_grid=dict(learning_rate=[0.01, 0.1, 0.3, 0.5]),
                             scoring='neg_mean_absolute_error',
                             refit=True,
                             cv=5)

xgb_reg_model.fit(X, y_spending)
```

```python
print("Best score: ", xgb_reg_model.best_score_)
print("Best params: ", xgb_reg_model.best_params_)
print("Best estimator: \n", xgb_reg_model.best_estimator_)

predictions_reg = xgb_reg_model.predict(X)
```

```python
# Machine learning (classification)

xgb_clf = XGBClassifier(objective='binary:logistic', random_state=123)

xgb_clf_model = GridSearchCV(estimator=xgb_clf, 
                             param_grid=dict(learning_rate=[0.01, 0.1, 0.3, 0.5]),
                             scoring='roc_auc',
                             refit=True,
                             cv=5)

y_prob = features['spend_90_flag']

xgb_clf_model.fit(X, y_prob)
```

```python
print("Best score: ", xgb_clf_model.best_score_)
print("Best params: ", xgb_clf_model.best_params_)
print("Best estimator: \n", xgb_clf_model.best_estimator_)

predictions_clf = xgb_clf_model.predict_proba(X)
```

```python
# Feature important

imp_spend_amount_dict = xgb_reg_model \
    .best_estimator_ \
    .get_booster() \
    .get_score(importance_type = 'gain') 

imp_spend_amount_dict
```

```python
imp_spend_prob_dict = xgb_clf_model \
    .best_estimator_ \
    .get_booster() \
    .get_score(importance_type = 'gain') 

imp_spend_prob_dict
```

```python
pd.DataFrame(predictions_clf)
```

```python
# Save Predictions
predictions_df = pd.concat(
    [
        pd.DataFrame(predictions_reg).set_axis(['pred_spend'], axis=1),
        pd.DataFrame(predictions_clf)[[1]].set_axis(['pred_prob'], axis=1),
        features.reset_index()
    ], 
    axis=1
)

predictions_df
```

```python
# Answering questions:
# 1. Which customers have the highest spend probability in next 90-days? 
#     - Target for new products similar to what they have purchased in the past

predictions_df.sort_values('pred_prob', ascending=False)
```

```python
# 2. Which customers have recently purchased but are unlikely to buy? 
#    - Incentivize actions to increase probability
#    - Provide discounts, encourage referring a friend, nurture by letting them know what's coming

predictions_df \
    [
        (predictions_df['recency'] < -90) & (predictions_df['pred_prob'] < 0.50)
    ] \
    .sort_values('pred_prob', ascending=False)
```

```python
# 3. Missed opportunities: Big spenders that could be unlocked ----
#    - Send bundle offers encouraging volume purchases
#    - Focus on missed opportunities

predictions_df \
    [
        predictions_df['spend_90_total'] == 0.0
    ] \
    .sort_values('pred_spend', ascending=False) 
```
